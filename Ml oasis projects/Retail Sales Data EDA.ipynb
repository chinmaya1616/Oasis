# 4.Customer and Product Analysis: Analyze customer demographics and purchasing behavior.

```python
#for gender distribution
sns.countplot(data = rsd,x='Gender')
plt.title('Gender Distribution')
plt.show()
```

```python
gender_counts = rsd['Gender'].value_counts()
gender_counts
```

```python
plt.figure(figsize=(6,6))
plt.pie(gender_counts,labels = gender_counts.index,autopct = '%1.1f%%')
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
#to know age range
sns.histplot(rsd['Age'],bins = 20)
plt.show()
```

```python
product_counts = rsd['Product Category'].value_counts()
product_counts
```

```python
plt.figure(figsize=(6,6))
plt.pie(product_counts,labels = product_counts.index,autopct = '%1.1f%%')
plt.title('Product Distribution')
plt.axis('equal')
plt.show()
```

```python
age_bins = [0,18,25,35,45,55,65,100]
age_labels = ['0-17','18-24','25-34','35-44','45-54','55-64','65+']
rsd['Age_group'] = pd.cut(rsd['Age'],bins=age_bins,labels=age_labels,right=False)
```

```python
#to check which age group buys from which product category in how much quantity
grouped_1 = rsd.groupby("Product Category")[["Quantity","Total Amount"]].sum()
grouped_1
```

```python
grouped_2 = rsd.groupby(["Product Category","Age_group"])[["Quantity","Total Amount"]].sum()
grouped_2
```
