Under Sampling - 
1. Build sample dataset containing similar distribution of legit transaction and fraudulent transaction
Number of fraud trans = 492

```python
legitsample = legit.sample(n=492) #random sampling
```

```python
#concatenating two dataframes
```

```python
new_dataset = pd.concat([legitsample,fraud],axis=0) #axis =0 data frames added one by one row wise if 1 then column wise added
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting data into Features and targer

```python
X = new_dataset.drop(columns = 'Class' ,axis = 1) #dropping columns
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

```python
#splitting into training and testing data
```

```python
X_train,X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify = Y,random_state = 2)
```

```python
X.shape
```

```python
X_train.shape #80%
```

```python
X_test.shape #20%
```

Model Training
Logistic regression for binary classification

```python
model = LogisticRegression()
```

```python
#training the modelwith training data
```

```python
model.fit(X_train,Y_train) #X_train = features Y_train = corresponding labels
```

```python
#Model evaluation based on accuracy score
```

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train) #compare values predicted
```

```python
print('Accuracy on training data: ',training_data_accuracy)
```

```python
#92.3% is good accuracy score
```

```python
#Accuracy score on test data(imp) 
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score of test data: ',test_data_accuracy)
```

```python
#accuracy score is 89.3% nearly same to training data 
```

```python
#overfitting and underfitting has not happened
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mse = mean_squared_error(X_test_prediction,Y_test)
r2 = r2_score(X_test_prediction,Y_test)

print('mse',mse)
print('r2',r2)
```
